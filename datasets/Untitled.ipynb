```python
import feature_analysis
import pandas as pd
import numpy as np
```

```python
fish = pd.read_csv("fish_test.csv")
```

```python
fish
```

```python
report = feature_analysis.regression.reg_analysis(fish, "Weight", "Date")
```

```python
report.analyse()
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
a = report.features_df
```

```python
df = feature_analysis.binary_class.feature_types.fill_missing(a, "mean")
df = feature_analysis.binary_class.feature_types.labelled_df(df)
```

```python
mae = []
for column in df.columns:
    rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
    rf.fit(df[[column]], report.y)
    # Calculate the absolute errors
    predictions = rf.predict(df[[column]])
    errors = abs(predictions - report.y)
    # Print out the mean absolute error (mae)
    mae.append(round(np.mean(errors), 2))
```

```python
summary = report.analyse()
```

```python
summary["MAE"] = mae
```

```python
summary.sort_values("MAE")
```
