```python
import pandas as pd
import datetime
pd.options.mode.chained_assignment = None 

## Data Preprocessing:
weather = pd.read_csv("weatherAUS.csv")
weather.Date = pd.to_datetime(weather.Date)
weather_test = weather.loc[[not x for x in weather["RainTomorrow"].isnull()]]

def transform_date(x):
    if x < datetime.datetime(2012, 1, 1): 
        return "1. Before 2012"
    elif x < datetime.datetime(2015, 1, 1): 
        return "2. After 2012, Before 2015"
    else:
        return "3. After 2015"
    
weather_test.Date = [transform_date(x) for x in weather_test.Date]
```

```python
import feature_analysis
```

```python
report = feature_analysis.binary_class.feature_analysis(weather_test, "RainTomorrow", "Date", bad_class="Yes")
```
